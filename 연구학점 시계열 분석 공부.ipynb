```python
Q1 = kodex200[['Volume']].quantile(q=0.25)
Q3 = kodex200[['Volume']].quantile(q=0.75)
print(Q1)
print(Q3) #이상치 제거 할 때 IQR을 기반으로 이상치를 탐지하는 예시 기반 이상치 확인 및 제거 방법 선택함 
```

```python
IQR = Q3-Q1
IQR #4분범위 구해줌 
```

```python
IQR_kodex200 = kodex200[(kodex200['Volume'] <= Q3['Volume']+1.5*IQR['Volume']) & (kodex200['Volume'] >= Q1['Volume']-1.5*IQR['Volume'])]

print('IQR_kodex200 :', len(IQR_kodex200)) #3사분위수 + 1.5 * 4분범위 보다 작은 값 그리고 1사분위수 - 1.5*4분범위 보다 큰 값만 변수에 저장 
```

```python
def adf_test(x):
    stat, p_value, lags, nobs, crit, icb = adfuller(x) #adfuller 함수를 이용하면 6가지 output이 나옵니다
    print('ADF statistics')
    print(stat) #ADF 검정 통계량입니다
    print('P-value')
    print(p_value)
    #두개만 프린트해본다 
```

```python
def adf_test(x):
    stat, p_value, lags, nobs, crit, icb = adfuller(x) #adfuller 함수를 이용하면 6가지 output이 나옵니다
    print('ADF statistics')
    print(stat) #ADF 검정 통계량입니다
    print('P-value')
    print(p_value) # 다시 한번 adf해주고 ~ 
```

```python
adf_test(data_transformed) #1차 차분후 다시 adf검정 
```

```python
mydata = kodex200[["Open", 'Close']]
mydata.index = kodex200["Volume"]
mydata.head() #다른 방법과 다른 컬럼으로 adf 검정 해보고 
```

```python
adfuller_test = adfuller(mydata['Open'], autolag= "AIC")
print("ADF test statistic: {}".format(adfuller_test[0]))
print("p-value: {}".format(adfuller_test[1])) 
#그냥 프린트를 빼주면 코드도 간단하고 format은 소수점 할 떄 쓰는 거 앞에는 그 문자 프린트 해주는 중
```

```python
adfuller_test = adfuller(mydata['Close'], autolag= "AIC")
print("ADF test statistic: {}".format(adfuller_test[0]))
print("p-value: {}".format(adfuller_test[1])) #이거는 open이랑 close로 상관분석한 거 
```
